# Data Ingestion Stage

This notebook mirrors the first stage of the project pipeline: download the dataset from Google Drive, extract it into the `artifacts` directory, and inspect the class distribution.

```python
from pathlib import Path
import zipfile
import gdown

DATA_URL = "https://drive.google.com/file/d/1T0-2_nce0eaLfeq-jpVdsJjVpz3rdbhh/view?usp=sharing"
DATA_DIR = Path("artifacts/data_ingestion")
ZIP_PATH = DATA_DIR / "data.zip"
EXTRACT_DIR = DATA_DIR / "Chest-CT-Scan-data"

DATA_DIR.mkdir(parents=True, exist_ok=True)
DATA_DIR, ZIP_PATH, EXTRACT_DIR
```

```python
if not ZIP_PATH.exists():
    gdown.download(DATA_URL, str(ZIP_PATH), quiet=False, fuzzy=True)
else:
    print(f"Dataset archive already exists at {ZIP_PATH}")
```

```python
if not EXTRACT_DIR.exists():
    with zipfile.ZipFile(ZIP_PATH, "r") as zip_ref:
        zip_ref.extractall(DATA_DIR)
    print("Extraction completed")
else:
    print(f"Dataset already extracted at {EXTRACT_DIR}")
```

```python
class_counts = {}
for class_dir in sorted(EXTRACT_DIR.iterdir()):
    if class_dir.is_dir():
        class_counts[class_dir.name] = len([file for file in class_dir.iterdir() if file.is_file()])
class_counts
```
